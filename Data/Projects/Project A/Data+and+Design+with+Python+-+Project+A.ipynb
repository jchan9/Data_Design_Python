The dataset displayed is of the most popular baby names in New York City. The data is based off of the baby's sex and the mother's ethnicity. There are multiple variables that are included. Here is a breakdown of them:

_____________
**Year of Birth**: 

* What year the baby is born. This dataset has information from years ranging from 2011-2014.

**Gender**:

* Whether the baby is male or female can contribute largly to what name is being given.

**Ethnicty**: 

* This is **NOT** the ethnicity of the baby, but of the mother. 

**Child's First Name**: 

* This is the name given to the child that is born.

**Count**: 

* This indicates how many children are named whatever is under ***Child's First Name***, given that the following are the same: ***Year of birth***, ***Gender***, and ***Ethnicity***(of mother).

**Rank**: 

* This indicates the popularity of the name given to the baby

# Slice Data by Row

_______

# Using .iloc

```python
names.iloc[5:26]
```

***Here we are able to see all the data from row #5-25***

\*_Notice that we cannot see row #26, even though we typed in from 5-26 in the command. The reason is because using ***.iloc*** will give us the data from one less than the row # we want to end at. If we typed in ***.iloc[10:20]*** we would get data from the rows numbered 10-19._\*

# Using .loc

```python
names.loc[5:26]
```

***Here we are able to see data from the rows numbered 5-26***

\*_Notice how this table has data from the row numbered 26. The same numbers were punched into the command, so why would this one show the row numbered 26 while the previous one didn't? The answer is simple, we may have typed in the same numbers into the command, but the difference is that the commands have changed. The previous one used ***.iloc***, which only gave us data from 5-25. By using ***.loc***, we are able to get data from 5-26._\*

# Using .loc on Variables

```python
names.loc[gender=="FEMALE"]
```

***Here we can sort out the table to show only the data for females that were born from 2011-2014***

\*_It is very important to note that using ***.iloc*** will ***NOT*** work on variables that are ***NOT*** numbers due to the fact that ***.iloc*** gives you the data of one less than what you type in. With variables that are not numbers, you cannot have one less than what you type in. If you typed in ***56***, ***.iloc*** would give you ***55***, but if you typed in ***female***, ***.iloc*** would give you an error because what is one less than female? It is nonexistant._\*

# Distplot

**Number of boys born vs. Number of girls born between 2011-2014**

```python
male = names.loc[gender=="MALE", ['Gender','Count']]
female = names.loc[gender=="FEMALE", ['Gender','Count']]
```

```python
male.head()
```

```python
sns.distplot(male["Count"])
```

```python
female.head()
```

```python
sns.distplot(female["Count"])
```

```python
sns.distplot(male["Count"])
sns.distplot(female["Count"])
```

We are able to compare the data between girls and boys to see the amount of times each name was being given. By looking at the count and adding those numbers up for each gender, we are able to get a sense of how many boys and girls were born between the years 2011-2014. When we put it onto a chart, we are able to see how the two populations of boys and girls overlap.

***Here we are able to see the sum of all males born in NYC from the years 2011-2014***

```python
sumMalesBorn = names.loc[gender=='MALE', 'Count'].sum()
```

```python
sumMalesBorn
```

***Here we are able to see the sum of all females born in NYC from the years 2011-2014***

```python
sumFemalesBorn = names.loc[gender=='FEMALE', 'Count'].sum()
```

```python
sumFemalesBorn
```

It is clear that 428,785 is larger than 331,705 - therefore the total birthrate of boys was larger than the total birthrate of girls between the years 2011-2014.

\*\*\*Note that this means that the amount of boys born within those 4 years was larger than that of the girls. It does **NOT** mean that the amount of boys born each year was larger than that of the girls. Some years there might have been more girls than boys born, but overall it was the boys that were born more frequently\*\*\*
